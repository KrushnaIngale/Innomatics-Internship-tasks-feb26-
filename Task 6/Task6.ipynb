**Problem Statement 1:Smart Parking Lot Management System**

Design a function to manage a smart parking lot.

The system should:

Accept vehicle entry and exit logs


Calculate total parked vehicles


Identify peak parking usage


Alert if parking exceeds capacity


```python
Parking_Capacity = 50
Vehicle_Logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

def calculate_parked_vehicles(Parking_Capacity,Vehicle_Logs):
  """Calculates the number of currently parked vehicles and their status."""
  parked_vehicles = 0 # Initialize parked_vehicles

  for log in Vehicle_Logs:
      if log == "IN":
          parked_vehicles += 1
      elif log=="OUT":
          parked_vehicles -= 1

  print(f"Currently parked vehicles: {parked_vehicles}")
  print("Parking Status: Available" if parked_vehicles < Parking_Capacity else "Parking Status: Unavailable")

calculate_parked_vehicles(Parking_Capacity,Vehicle_Logs)
```

 **Problem Statement 2:Online Food Delivery Time Estimator**

Create a function that estimates delivery time based on:

Distance (km)


Weather condition


Traffic level


Apply delays dynamically and display final ETA.


```python
distance = 8
traffic = "High"
weather = "Rainy"

def estimate_delivery_time(distance_km, weather_condition, traffic_level):
    base_time_per_km = 5  # minutes per km
    estimated_time = distance_km * base_time_per_km

    # Add delay based on weather
    if weather_condition == "Rainy":
        estimated_time += 5  # Add 5 minutes for rainy weather
    elif weather_condition == "Snowy":
        estimated_time += 15 # Add more for snowy weather

    # Add delay based on traffic
    if traffic_level == "High":
        estimated_time += 10 # Add 10 minutes for high traffic
    elif traffic_level == "Medium":
        estimated_time += 5  # Add 5 minutes for medium traffic

    return estimated_time

# Test with sample input
estimated_eta = estimate_delivery_time(distance, weather, traffic)
print(f"Estimated Delivery Time: {estimated_eta} minutes")
```

 **Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**

Build a function that analyzes seat booking data and:

Calculates occupancy percentage


Determines if show is Housefull


Suggests opening additional shows


```python
total_seats=200
booked_seats=[1] * 150 # Assuming the '150 entries'

def analyze_seat_occupancy(total_seats, booked_seats):
    occupied_seats = len(booked_seats)
    occupancy_percentage = (occupied_seats / total_seats) * 100

    print(f"Occupancy: {occupancy_percentage:.0f}%") # Format as a whole number

    if occupancy_percentage == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open additional shows!")
    elif occupancy_percentage >= 75: # Assuming 75% or more is 'Almost Full' and warrants suggestion
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional shows.")
    else:
        print("Show Status: Available") # Default status for lower occupancy

analyze_seat_occupancy(total_seats, booked_seats)
```

**Problem Statement 4:Cloud Server Load Classification System**

Create a function to classify server load based on CPU usage readings.
Rules:

Average CPU < 50% → Normal


50%–80% → Warning


80% → Critical


```python
cpu_readings=[45, 60, 70, 85, 90]

def classify_server_load(cpu_readings):
    """Classifies server load based on the average CPU usage readings."""
    total_readings = len(cpu_readings)
    # Calculate the average CPU load from the given readings
    avg_cpu_load=sum(cpu_readings)/total_readings

    print(f"Average CPU Load: {avg_cpu_load:.0f}%")
    # Classify the server status based on the average CPU load
    if avg_cpu_load < 50:
        print("Server status : Normal") # CPU load below 50% is considered Normal
    elif 50 <= avg_cpu_load <=80:
        print("Server status : Warning") # CPU load between 50% and 80% is a Warning
    else:
        print("Server status : Critical") # CPU load above 80% is Critical

classify_server_load(cpu_readings)
```

 **Problem Statement 5:Smart Classroom Resource Usage Monitor**

Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.


```python
resource_usage = {
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}

def monitor_resource_usage(resource_usage, usage_threshold=7): # Added a default threshold of 7 hours
    """Tracks usage of classroom resources and identifies overuse patterns."""
    overused_resources = []
    # Iterate through each resource and its usage hours
    for resource, usage_hours in resource_usage.items():
        # Check if usage exceeds the defined threshold
        if usage_hours > usage_threshold:
            overused_resources.append(resource)

    # Based on the identified overused resources, print the status and energy alert
    if overused_resources:
        print(f"Overused Resources: {', '.join(overused_resources)}")
        print("Energy Alert: Yes") # Alert if any resources are overused
    else:
        print("Overused Resources: None")
        print("Energy Alert: No") # No alert if all resources are within limits

# Call the function with the sample input
monitor_resource_usage(resource_usage)
```

**Problem Statement 6: Online Event Registration Capacity Controller**

Create a function that manages event registrations by:

Tracking registrations


Preventing overbooking


Triggering waitlist mode


```python
event_capacity= 100
registrations=105

def manage_event_registrations(event_capacity,registrations):
  """Manages event registrations, prevents overbooking, and triggers waitlist mode."""
  # Check if current registrations are within the event's capacity
  if registrations < event_capacity:
    print(f"Confirmed Registrations: {registrations}")
    print("Registration Status: Open") # Event is still open for registrations
  else:
    # If registrations exceed capacity, confirm up to capacity and calculate waitlisted users
    print(f"Confirmed Registrations: {event_capacity}")
    print(f"Waitlisted Users: {registrations - event_capacity}")
    print("Registration Status: Closed") # Event is full, new registrations go to waitlist

manage_event_registrations(event_capacity,registrations)
```
