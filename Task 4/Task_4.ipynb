**Problem 1: Mobile Recharge Validation System**

 Problem Statement
Create a function to validate a mobile recharge request.

Rules:


*   Recharge amount must be ≥ ₹50
*   Valid plans: 199, 299, 399, 599
*   User should be allowed to retry if invalid











```python
# Defines a function to validate mobile recharge requests
def validate_recharge(amount, plan):
  # Checks if the amount is at least 50 and the plan is one of the valid options
  if(amount>=50 and plan in [199,299,399,599]):
    return True
  else:
    return False

# Loop to allow user to retry recharge until successful
while(True):
  # Prompts user for recharge amount and plan
  amount=int(input("Enter the amount: "))
  plan=int(input("Enter the plan: "))
  # Calls the validation function and prints success/failure message
  if(validate_recharge(amount,plan)):
    print("Recharge successful")
    break # Exits loop on successful recharge
  else:
    print("Recharge failed...Retry: ") # Prompts retry on failure
```

**Problem 2: Inventory Reorder Alert System**

Problem Statement
Given a dictionary of products and stock quantities:

* Print Reorder Alert if stock < 15


* Print Stock OK otherwise


```python
# Dictionary storing product stock quantities
products={'A':20 , 'B':10, 'C':13,'D':22}

# Defines a function to check stock levels and print reorder alerts
def reorder_alert(products):
  # Iterates through each product and its stock quantity
  for key,value in products.items():
    # Prints 'Reorder Alert' if stock is less than 15
    if(value<15):
      print(key,"Reorder Alert")
    # Prints 'Stock OK' otherwise
    else:
      print(key,"Stock OK")

# Calls the reorder_alert function with the products dictionary
reorder_alert(products)
```

**Problem 3: Student Result Processing System**

Problem Statement
Create a function that:

* Accepts a list of student marks


* Calculates average marks


* Prints Pass if average ≥ 50 else Fail

```python
# List of student marks
marks=[69,73,77,92,95,99]

# Defines a function to process student results based on marks
def student_result(marks):
  # Calculates the average of the marks
  avg=sum(marks)/len(marks)
  # Returns "Pass" if the average is 50 or more
  if(avg>=50):
    return ("Pass")
  # Implicitly returns None if average is less than 50 (Fail condition not explicitly returned)

# Prints the result of the student_result function
print(student_result(marks))
```

**Problem 4: Cab Fare Estimator with Retry Option**

Problem Statement
Create a cab fare calculator:

* Base fare = ₹50


*  ₹12 per km


* Peak hour → 25% extra


* Allow retry using while loop


```python
# Defines a function to estimate cab fare
def cab_fare_estimator(distance,peak_hour):
  base_fare=50 # Base fare for the cab ride
  fare_per_km=12 # Fare per kilometer
  fare=base_fare+(distance*fare_per_km) # Calculate initial fare
  # Applies 25% extra charge if it's a peak hour
  if(peak_hour):
    fare=fare*1.25
  return fare

# Loop to allow user to calculate multiple fares with a retry option
while(True):
  try:
    # Prompts user for distance and checks if it's a peak hour
    distance=int(input("Enter the distance: "))
    peak_hour_input=input("Is it a peak hour? (y/n): ")
    peak_hour= (peak_hour_input.lower()=='y') # Converts input to boolean
    # Prints the calculated fare
    print("The fare is: ",cab_fare_estimator(distance,peak_hour))

    # Asks user if they want to calculate another fare
    retry = input("Do you want to calculate another fare? (y/n): ")
    if retry.lower() != 'y':
      print("Thank you...") # Exits loop if user doesn't want to retry
      break
  except ValueError:
    # Handles invalid input for distance
    print("Invalid input. Please enter a valid number for distance.")
```

**Problem 5: Employee Attendance Eligibility Checker**

Problem Statement
Create a function that:

* Accepts attendance list ("P", "A")


* Calculates attendance percentage


* Returns Eligible if ≥ 75% else Not Eligible

```python
# List representing employee attendance ('P' for Present, 'A' for Absent)
attendance_list=['P','P','P','A','P','P','P','P','P','P']

# Defines a function to check employee attendance eligibility
def employee_attendance_checker(attendance_list):
  total_days=len(attendance_list) # Total number of days
  present_days=attendance_list.count('P') # Number of present days
  attendance_percentage=(present_days/total_days)*100 # Calculate attendance percentage
  # Returns "Eligible" if attendance is 75% or higher, "Not Eligible" otherwise
  if(attendance_percentage>=75):
    return "Eligible"
  else:
    return "Not Eligible"

# Prints the eligibility status of the employee
print(employee_attendance_checker(attendance_list))
```

**Problem 6: Password Strength Checker**

Problem Statement
Create a function to check password strength.

Rules:

* Minimum length = 8


* Must contain at least one digit


* Must contain at least one special character (@#$)

```python
# Defines a function to check the strength of a password
def passWord_strength_checker(password):
  pw_len = len(password) # Get password length
  has_digit = any(char.isdigit() for char in password) # Check for at least one digit
  # Check for at least one special character (@, #, or $)
  has_special = any(char in ['@', '#', '$'] for char in password)

  # Password is 'Strong' if it meets all criteria
  if pw_len >= 8 and has_digit and has_special:
    return "Strong"
  # Otherwise, it's a 'Weak PW'
  else:
    return "Weak PW"

# Loop to continuously prompt for password until a strong one is entered
while(True):
  password=input("Enter the password: ")
  # Checks if the entered password is strong
  if(passWord_strength_checker(password)=="Strong"):
    print("Password is strong ")
    break # Exits loop if password is strong
  else:
    print("Password is weak, please try again...") # Prompts retry if password is weak
```
