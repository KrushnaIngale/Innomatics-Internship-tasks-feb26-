**Problem 1: Social Media – Post Engagement Analyzer**

Problem Statement

Create a function to analyze engagement on a social media post.

Rules

Calculate total likes


If Total Likes ≥ 1000 → Viral Post

Otherwise → Normal Engagement


```python
def analyze_post_engagement(total_likes):

    # Check if the total likes meet the viral post threshold
    if total_likes >= 1000:
        return "Viral Post" # Return 'Viral Post' if likes are 1000 or more
    else:
        return "Normal Engagement" # Otherwise, return 'Normal Engagement'

# Read input: a space-separated list of integers representing individual likes
post_data = list(map(int, input().split()))

# Initialize total_likes counter
total_likes = 0
# Iterate through the list of likes and sum them up
for likes in post_data:
    total_likes += likes

# Call the function to analyze engagement with the calculated total likes
engagement = analyze_post_engagement(total_likes)
# Print the final total likes and the post status based on engagement analysis
print(f'Total Likes: {total_likes}\nPost Status: {engagement}')
```

**Problem 3: Agriculture – Rainfall Adequacy Checker**

Problem Statement

Create a function that analyzes seasonal rainfall data.

Rules

Calculate average rainfall

Average ≥ required level → Adequate Rainfall

Otherwise → Inadequate Rainfall


```python
def check_rainfall_adequacy(rainfall_data, required_level):
    # Check if rainfall data is provided to avoid division by zero
    if not rainfall_data:
        return "No rainfall data provided."

    # Calculate the average rainfall from the given data
    average_rainfall = sum(rainfall_data) / len(rainfall_data)
    # Print the calculated average rainfall, formatted to zero decimal places
    print(f"Average Rainfall: {average_rainfall:.0f}")

    # Compare the average rainfall to the required level to determine adequacy
    if average_rainfall >= required_level:
        return "Rainfall Status: Adequate Rainfall"
    else:
        return "Rainfall Status: Inadequate Rainfall"


# Example 1: Sufficient rainfall scenario
rainfall_measurements = [60, 75, 80, 70, 75]
required_min_rainfall = 72
print(check_rainfall_adequacy(rainfall_measurements, required_min_rainfall))

# Example 2: Insufficient rainfall scenario
rainfall_measurements_low = [40, 50, 30, 45]
required_min_rainfall_low = 60
print(check_rainfall_adequacy(rainfall_measurements_low, required_min_rainfall_low))
```

**Problem 4: Social Media – Duplicate Account Detection**

Problem Statement

Create a function to detect duplicate usernames in a system.

Rules


If duplicates exist → Flag accounts
Else → Accounts are unique


```python
def detect_duplicate_accounts(usernames):
  # Create a set from the list of usernames to get only unique usernames.
  # Sets automatically handle uniqueness, so duplicates are removed.
  unique_usernames = set(usernames)

  # Compare the length of the original list with the length of the set.
  # If they are not equal, it means there were duplicates in the original list.
  if len(unique_usernames) != len(usernames):
      return "Duplicate Accounts Found: Yes" # Return 'Yes' if duplicates are found
  else:
    return "Duplicate Accounts Found: No" # Return 'No' if all accounts are unique


# Example usage:
usernames = ["abc123","pqr123","abc1234"]

# Call the function and print the result
print(detect_duplicate_accounts(usernames))
```

**Problem 6: Agriculture – Premium Crop Price Filter**

Problem Statement

Create a function that filters crop prices above a premium threshold.

Rules

Price > ₹2000 → Premium Crop
Else → Regular Crop


```python
def filter_crop_prices(crop_prices):
  premium_crops = []
  for price in crop_prices:
    if price > 2000:
      premium_crops.append(price)
  return premium_crops

crop_prices = [1800,2000,1300,2500,2900]
print("Premium crops: ",filter_crop_prices(crop_prices))
```

**Problem 9: E-Learning – Student Attendance Eligibility System**

Problem Statement

Create a function to calculate attendance percentage and check exam eligibility.

Rules

Attendance ≥ 75% → Eligible
Attendance < 75% → Not Eligible


```python
def check_attendance_eligibility(attendance_list):
    # Calculate the total number of days recorded in the attendance list
    total_days = len(attendance_list)
    # Initialize a counter for days the student was present
    present_days = 0

    # Iterate through each entry in the attendance list
    for day in attendance_list:
        # If the entry indicates 'P' (Present), increment the present_days counter
        if day == "P":
            present_days += 1

    # Calculate the attendance percentage
    attendance_percentage = (present_days / total_days) * 100 if total_days > 0 else 0

    # Check if the attendance percentage meets the eligibility threshold (75%)
    if attendance_percentage >= 75:
        return f"Attendance Percentage: {attendance_percentage}\nExam Eligibility: Eligible"
    else:
        return f"Attendance Percentage: {attendance_percentage}\nExam Eligibility: NOt Eligible"

# Example usage: Define an attendance list with 'P' for present and 'A' for absent
attendance_list = ["P", "P", "A", "P", "P"]

print(check_attendance_eligibility(attendance_list))
```

**Problem 10: Smart Electricity Bill Analyzer**

Problem Statement

You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.

Billing Rules

First 100 units → ₹3 per unit
Next 100 units (101–200) → ₹5 per unit
Above 200 units → ₹7 per unit

Usage Classification

Total bill < ₹500 → Low Usage
₹500 – ₹1500 → Moderate Usage
₹1500 → High Usage


```python
def calculate_electricity_bill(units):
  bill = 0
  # Calculate bill based on units consumed using a tiered pricing structure
  if units > 200:
      # For units above 200, apply the highest rate to the excess units
      # and standard rates to the first 200 units
      bill = 100 * 3 + 100 * 5 + (units - 200) * 7
  elif units > 100 and units <= 200:
      # For units between 101 and 200, apply the second tier rate to excess units
      # and the first tier rate to the first 100 units
      bill = 100 * 3 + (units - 100) * 5
  elif units > 0 and units <= 100:
      # For units up to 100, apply the lowest tier rate
      bill = units * 3

  # Classify usage level based on the total bill amount
  if bill < 500:
      return f"Total Bill: {bill}\nUsage Classification: Low Usage"
  elif bill >= 500 and bill <= 1500:
      return f"Total Bill: {bill}\nUsage Classification: Moderate Usage"
  else:
      return f"Total Bill: {bill}\nUsage Classification: High Usage"

# Get units consumed from user input
units = int(input("Enter the units: "))
# Calculate and print the electricity bill and usage classification
print(calculate_electricity_bill(units))
```
