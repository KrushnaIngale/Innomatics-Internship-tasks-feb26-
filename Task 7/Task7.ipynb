**Problem 1: Smart Light Controller**

Problem Statement

Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # Default state

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} light is {self.status}")


# Sample usage
light_name = input("Light Name: ")
action = input("Action: ")

light = SmartLight(light_name)

if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

light.display_status()
```

**Problem 2: Employee ID Card System**

 Problem Statement

Create a class that stores employee name, ID, and department and displays ID card details.


```python
class EmployeeIDCard:
    def __init__(self, name, id, department):
        self.name = name
        self.id = id
        self.dept = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        print(f"Department: {self.dept}")

emp = EmployeeIDCard("Prerna", "EMP137", "IT")
emp.display_id_card()
```

**Problem 3: Mobile Contact Record**

Problem Statement

Create a class to store contact name and phone number and display contact information.


```python
class MobileContact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

contact=MobileContact("Prerna", "9876543210")
contact.display_contact()
```

**Problem 4: Product Price Tag Generator**

Problem Statement

Create a class that stores product name and price and prints a formatted price tag.


```python
class ProductPriceTag:
  def __init__(self, name, price):
    self.name = name
    self.price = price
  def display_price_tag(self):
    print(f"Product: ", self.name)
    print(f"Price: ₹", self.price)

product_name="Headphones"
price=2499
product = ProductPriceTag(product_name,price )
product.display_price_tag()
```

**Problem 5: Movie Rating Display System**

Problem Statement

Create a class that stores movie name and rating and displays movie details.
Real-Time Use


```python
movie_name="Avengers- Doomsday"
ratings=4.9

class MovieRating:
  def __init__(self,movie_name,rating):
    self.movie_name=movie_name
    self.rating=rating
  def display_movie_details(self):
    print(f"Movie: {self.movie_name}")
    print(f"Rating: {self.rating}/5")

movie=MovieRating(movie_name,ratings)
movie.display_movie_details()
```
