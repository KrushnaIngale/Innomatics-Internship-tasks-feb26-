1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"

Requirements:
Print "Login Successful" if both username and password match
Otherwise print "Invalid Credentials"
Real-World Application: Authentication systems


```python
username = "admin"
password = "1234"

user = input("Enter your username: ")
pw = input("Enter your password: ")

if user == username and pw == password:      #logical checking
    print( "Login Successful")
else:
    print("Invalid Credentials")
```

2. Pass / Fail Analyzer
Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:
A student passes if marks â‰¥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly
Real-World Application: Academic evaluation systems


```python
marks = [45, 78, 90, 33, 60] # Marks of indivisual students are in the list
totalCounts=0
totalFails=0
for i in marks:
    if i >= 50:
        totalCounts+=1
    else:
        totalFails+=1
print("Total Pass: ",totalCounts)
print("Total Fail: ",totalFails)
```

3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

Requirements:
Remove extra spaces
Convert all names to lowercase
Print the cleaned list
Real-World Application: Data preprocessing before analysis


```python
names = [" Alice ", "bob", " CHARLIE "]
def cleanData(names):
  cleaned=[]
  for i in names:
    # cleaning the data
    data=i.strip()
    data=data.lower()
    cleaned.append(data)
  return cleaned
print("Cleaned Data: ",cleanData(names))
```

Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

Print the length of each message

```python
messages = ["Hi", "Welcome to the platform", "OK"]

def msgAnalyzer(messages):
  count=0
  for i in messages:
    if len(i) > 10:
      print(f"This message '{i}' is flagged for being longer than 10 character: ", len(i))   # flagging for large messages
    else:
      print(f"Length of message: {i} is ",len(i))

msgAnalyzer(messages)
```

5. Error Message Detector
Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:
Count the number of "ERROR" entries
Print the total error count
Real-World Application: Monitoring and log analysis systems


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

def counrErrors(logs):
  errorCount=0;
  for i in logs:
    if i=="ERROR":
      errorCount+=1   #simple counter increment for exact match
  print("Total ERROR counts = ",errorCount)

counrErrors(logs)
```
